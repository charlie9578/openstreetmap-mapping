```python
class Tags(object):
    
    def __init__(self, *initial_data, **kwargs):
              
        for dictionary in initial_data:
            
            for key in dictionary:

                if isinstance(dictionary[key], dict):
                    if len(dictionary[key])>1:
                        setattr(self, 'value', key)
                        setattr(self, key, Tags(dictionary[key]))
                    else:
                        setattr(self, key, Tags(dictionary[key]))
                else:
                    setattr(self, key, dictionary[key])
            
    def __repr__(self):
        return self.value.replace("__",":")
    
    def __str__(self):
        return self.value.replace("__",":")
```

```python
df_guernsey = osm.toolkit.get_osm_data_by_id(member_id=3064410,obj_type="rel",output="geom") # Guernsey

df_guernsey_coast = pd.DataFrame.from_records(df_guernsey['members'][0])
```

```python
# a rather convoluted way to make sure the ways plot in the correct order
df_guernsey_coast.loc[:, 'first_coord'] = df_guernsey_coast['geometry'].map(lambda x: x[0])
df_guernsey_coast.loc[:, 'last_coord'] = df_guernsey_coast['geometry'].map(lambda x: x[-1])
```

```python
for row in df_guernsey_coast.iterrows():
    index = df_guernsey_coast[df_guernsey_coast['first_coord']==row[1]['last_coord']].index
    df_guernsey_coast.loc[row[0],'path']= int(index.values[0])
```

```python
df_guernsey_coast_ordered = pd.DataFrame()

for row in range(len(df_guernsey_coast)):
    
    #print(row)
    
    if row == 0:
        ordered = df_guernsey_coast.iloc[row]
        next_row = df_guernsey_coast.iloc[row]['path']
        
    else:
        ordered = df_guernsey_coast.iloc[int(next_row)]
        next_row = df_guernsey_coast.iloc[int(next_row)]['path']
        
    df_guernsey_coast_ordered = pd.concat([df_guernsey_coast_ordered,ordered],axis=1)    
```

```python
df_guernsey_coast_ordered = df_guernsey_coast_ordered.T
```

```python
coast_line = []

for row in df_guernsey_coast_ordered.iterrows():
    coast_line = coast_line + row[1]['geometry']

df_coast_line = pd.DataFrame.from_records(coast_line)
```

```python
# create the beaches
xs = []
ys = []
names = []

for row in df_beach.iterrows():
    way=pd.DataFrame.from_records(row[1]['geometry'])
       
    x,y = TRANSFORM_4326_TO_3857.transform(way['lat'],way['lon'])   
    
    names.append(row[1]['name'])
    xs.append(x)
    ys.append(y)
    
    
source_beach = ColumnDataSource({'xs':xs,
                                 'ys':ys,
                                 'name':names
                                 })    
```

```python
output_file("GuernseyBeaches.html")

# See https://wiki.openstreetmap.org/wiki/Tile_servers for various tile services
MAP_TILES = {"OpenMap": WMTSTileSource(url="http://c.tile.openstreetmap.org/{Z}/{X}/{Y}.png"),
         "ESRI": WMTSTileSource(url="https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{Z}/{Y}/{X}.jpg"),
         "OpenTopoMap": WMTSTileSource(url="https://tile.opentopomap.org/{Z}/{X}/{Y}.png")}

# Create a bokeh figure with tiles
plot_map = figure(plot_width=800, 
                  plot_height=800,
                  x_axis_type="mercator", 
                  y_axis_type="mercator",
                )

plot_map.add_tile(MAP_TILES['ESRI'])

tooltips = [("name", "@name")]

beaches = plot_map.patches(xs='xs',
                           ys='ys',
                           source=source_beach,
                           line_width=2,
                           line_color='yellow',
                           fill_color='yellow',
                           legend_label='beach')
    
labels = LabelSet(x='xs', 
                  y='ys', 
                  text='name',
                  source=source_beach,
                  text_font_size='16px',
                  render_mode='canvas')

plot_map.add_layout(labels)


x,y = TRANSFORM_4326_TO_3857.transform(df_coast_line['lat'],df_coast_line['lon'])   

source = ColumnDataSource({'x':x,
                           'y':y})

plot_map.patch(x="x", 
        y="y",      
        source=source,
        line_width=2,
        line_color='green',
        fill_color='white',
        legend_label='island')   


plot_map.add_tools(HoverTool(tooltips=tooltips, renderers=[beaches]))
```
